```python
#Reading the text file into a seperate variable

img_captions = 'C:/Users/parul_000/Downloads/Python/Capstone - Eye for the blind/archive/captions.txt'

def load_doc(filename):
    file = open(filename, 'r')
    text = file.read()
    file.close()
    return text

doc = load_doc(img_captions)
print(doc[:300])
```

We now need a dataframe which summarizes the image, path & captions

Each image id has 5 captions associated with it, hence the total dataset should have 40455 samples

```python
#Storing the image id, image path, and captions respectively in separate variables

all_img_id = []
all_img_vector = []
annotations = []

def getImageDetails(doc):
    doc = doc.splitlines()[1:] #adding [1:] this since we don't need the first row of headers
    for string in doc:
        comma = string.index(",")
        img_id = string[:comma]
        all_img_id.append(img_id)
        annotations.append(string[comma+1:])
        all_img_vector.append(images + "/" + img_id)

getImageDetails(doc)

df = pd.DataFrame(list(zip(all_img_id, all_img_vector,annotations)),columns =['ID','Path', 'Captions']) 
df.head()
```
